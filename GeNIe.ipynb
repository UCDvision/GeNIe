```python
from editor import ImageEditor
editor = ImageEditor()
```

```python
import PIL
from PIL import Image

filepath = 'dogs.JPG'
image = Image.open(filepath).convert('RGB')
image

```

```python
# Img2Img

percent_noise = 0.3 
prompt = 'A photo of a dog.'
output = editor.edit(image, prompt, percent_noise)
output
```
